```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from prettytable import PrettyTable
import re
import time


# Launch Chrome browser
driver = webdriver.Chrome()

# Open the LinkedIn search page
driver.get("https://www.linkedin.com/search/results/people/?currentCompany=%5B%221586%22%5D&geoUrn=%5B%22102713980%22%5D&network=%5B%22F%22%5D&origin=FACETED_SEARCH&sid=Ni3")

# Wait for user to fill in email and password
WebDriverWait(driver, 600).until(EC.url_changes(driver.current_url))
time.sleep(3)
```

```python
# Initialize an empty list to store all names
all_names = []

# Initialize cumulative count of names
cumulative_count = 0

# Loop through each page of search results
page_number = 1
while page_number <= 32:  # Fetch results till page number 2
    # Find all the name elements
    name_elements = driver.find_elements(By.XPATH, "//span[@aria-hidden='true']")
    time.sleep(3)  # Adding a small delay to ensure all elements are loaded

    # Extract names and store them in a list, filtering out non-alphabetic characters and excluding "Messaging"
    names_list = [element.text.strip() for element in name_elements if re.match("^[A-Za-z\s]+$", element.text.strip()) and element.text.strip() != "Messaging"]

    # Extend the list of all names with the current page's names
    all_names.extend(names_list)

    # Count the number of names
    names_count = len(names_list)

    # Update cumulative count of names
    cumulative_count += names_count

    # Print the page results
    print("\nPage Results (Page {}):".format(page_number))
    print("Names:", ", ".join(names_list))
    print("Count of Names:", names_count)
    print("Cumulative Count:", cumulative_count)

    # Scroll to the bottom of the page
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(3)  # Adding a delay to ensure the page scrolls to the bottom

    # Find the "Next" button
    next_button = None
    try:
        next_button = driver.find_element(By.XPATH, "//span[text()='Next']")
    except NoSuchElementException:
        break  # Exit the loop if the "Next" button is not found

    # Click the "Next" button to go to the next page
    next_button.click()
    time.sleep(3)  # Increasing the delay to ensure the next page is loaded

    # Increment page number
    page_number += 1

# Output the final summary
print("\nFinal Summary:")
print("Total Number of Pages:", page_number - 1)
print("All Names:", all_names)
print("Count of Names:", len(all_names))
print("Cumulative Count:", cumulative_count)
```
